## Jupyer 是什么
是一个及文本、代码、绘图于一身的轻量级的文本编辑器。它是基于iPython。

## 单元格
- 单元格常见有两个状态，选中状态和编辑状态。编辑状态可以看到光标在闪动。按<Esc>可以进入选中状态。可以删除单元格，新增单元格，选中上/下一个单元格等操作。
- 从选中状态可以按下 `<Enter>` 切换到编辑状态。

### 单元格的模式
单元格能够执行的功能。
- 代码模式：在这个模式下运行代码，运行指令`<Ctrl+Enter>`；
- Markdown模式：在这个模式下可以识别`Markdown`语法；
- 切换：在单元格的选中状态下，按下`<m>`markdown, `<y>`code；

### 操作
- 新增单元格：单元格处于选中状态（没有光标），按下`<b>`，下方新增。按下`<a>`上方新增；
- 删除当前单元格：单元格处于选中状态（没有光标），双击`<d>`，或者剪切`<x>`；
- 复制单元格：`<c>`；
- 粘贴单元格：`<v>`；
- 撤销操作：`<z>`；
